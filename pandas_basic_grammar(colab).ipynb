**(2) merge : 두 df를 열로 합치기**

```python
df1 = pd.DataFrame({"x1": ["한국", "미국", "중국"], "x2": [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({"x1": ["한국", "미국", "영국"], "x3": ["T", "F", "T"]})
df2
```

```python
# x1을 기준으로 df1과 df2를 합쳐보자
# default로 how=inner라 공통 행만 남는다.
pd.merge(df1, df2, on=['x1'])
```

```python
# df2 데이터 기준으로 df1을 합치고 싶다면?
pd.merge(df1, df2, on=['x1'], )
```

```python
# df1과 df2 모든 데이터를 살리며 합집합 하고 싶다면?
pd.merge(df1, df2, on=['x1'], )
```

```python
# 그럼 한 열말고 두 열을 기준으로 합치고 싶다면??
df3 = pd.DataFrame({"x1": ["한국", "미국", "중국"], "x2": ['사과','배','수박'], "x3": ["D", "F", "W"]})
df3
```

```python
df4 = pd.DataFrame({"x1": ["한국", "미국", "영국"], "x2": ['사과','오렌지','호박'], "x4": ['데이터', '만들기', '어렵다']})
df4
```

```python
# x1과 x2열 기준으로 합쳐본다.
pd.merge(df3, df4, on=[])
```

```python
# 각 행이 어떻게 merge된건지 출처를 알고 싶을 때 indicator를 씀
pd.merge(df3, df4, on=['x1', 'x2'], how='outer', indicator=True)
```
