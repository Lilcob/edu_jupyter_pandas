```
import numpy as np
import matplotlib.pyplot as plt
```

```
data = np.loadtxt(input_file, skiprows=1, delimiter=',')
data
```

```
x = data[:, 0]
y = data[:, 1]
x, y
```

```
plt.figure(figsize=(8, 8))
plt.plot(x, y, 'o')
```

```
def preprocessing(x):
  return (x - x.mean()) / x.std()
```

```
x = preprocessing(x)
x
```

```
def forward(x):
  yhat = w0 + w1 * x
  return yhat
```

```
w0 = np.random.rand()
w1 = np.random.rand()
w0, w1
```

```
plt.figure(figsize=(8, 8))
plt.plot(x, y, 'o')
plt.plot(x, forward(x))
```

```
def loss():
  return 1/len(y) * sum(forward(x) - y)**2
```

```
lr = 0.001
error = loss()
diff = 1
while diff > 0.01:
  w1 = w1 - lr*(2/len(y))*sum((forward(x)-y)*x)
  w0 = w0 - lr*(2/len(y))*sum(forward(x)-y)
  new_error = loss()
  diff = error - new_error
  print(f'cost:{new_error: .4f}, diff:{diff: .4f}')
  error = new_error
```

```
w1, w0
```

```
plt.figure(figsize=(10, 7))
plt.plot(x, y, 'o')
plt.plot(x, forward(x))
```

```
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
camera = celluloid.Camera(fig)  

lr = 0.01
error = loss()
diff = 1
while diff > 0.01:
  ax = plt.plot(x, y, 'o')
  ax = plt.plot(x, forward(x))
  camera.snap()   # 찰칵

  w1 = w1 - lr*(2/len(y))*sum((forward(x)-y)*x)
  w0 = w0 - lr*(2/len(y))*sum(forward(x)-y)
  new_error = loss()
  diff = error - new_error
  # print(f'cost:{new_error: .4f}, diff:{diff: .4f}')
  error = new_error
```
